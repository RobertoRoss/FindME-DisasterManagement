```python
from __future__ import print_function
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import sys
import sqlite3 as lite
from sklearn.metrics.pairwise import additive_chi2_kernel
from sklearn.metrics.pairwise import chi2_kernel
```

## LFW NUMBER OF IMAGES

```python
con = None

try:
    person_identifier = {}    # hash map for storing person ids
    con = lite.connect('../../db/facedb.litedb')
    
    cur = con.cursor()
    
    lfwCountQuery =  "select count(distinct id) from images;"                        
                    
    cur.execute(lfwCountQuery)
    lfwCountRes = cur.fetchall()
    lfwNumImages = lfwCountRes[0][0]
    
except lite.Error, e:
    print("Error %s:" % e.args[0])
    sys.exit(1)
    
finally:
    if con:
        con.close()
```

```python
 print('Number of Images :', lfwNumImages)
```

## LFW PAIR LABELS

```python
con = None

try:
    con = lite.connect('../../db/facedb.litedb')
    
    cur = con.cursor()
    
    lfwPairsLabelsQuery =  "select A.id, B.id, 1 " + \
                        "from " + \
                        "(select id, person_id from images) A, " + \
                        "(select id, person_id from images) B " + \
                        "where A.person_id = B.person_id " + \
                        "union " + \
                        "select C.id, D.id, 0 " + \
                        "from " + \
                        "(select id, person_id from images) C, " + \
                        "(select id, person_id from images) D " + \
                        "where C.person_id <> D.person_id;"
   
    lfwPairsLabels = {}    
    cur.execute(lfwPairsLabelsQuery)
    lfwPairsLabelsRes = cur.fetchall()  
    for t in lfwPairsLabelsRes:
        lfwPairsLabels[(t[0], t[1])] = t[2]
    
except lite.Error, e:
    print("Error %s:" % e.args[0])
    sys.exit(1)
    
finally:
    if con:
        con.close()
```

```python
print('Number of pairs: ', len(lfwPairsLabels))
```

```python
lfwPairsLabelsMat = np.zeros(shape=(lfwNumImages, lfwNumImages), dtype=np.float)
for k, v in lfwPairsLabels.iteritems():
    lfwPairsLabelsMat[k[0]-1, k[1]-1] = v
print(lfwPairsLabelsMat)
```

```python
fig, ax = plt.subplots()
lfwLabelsHMap = ax.pcolorfast(lfwPairsLabelsMat)
plt.show()
```
